```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('ds_salaries.csv')
data.head(5)
```

## 1. Средняя зарплата по вакансиям

```python
data.groupby(['job_title']).agg(['mean','median','count'])['salary_in_usd']
```

## 2. Средняя зарплата по регионам

```python
data.groupby(['company_location']).agg(['mean','median','count'])['salary_in_usd']
```

## 3. Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?

```python
data.groupby('job_title',as_index=False)[['salary_in_usd']].mean().sort_values(by='salary_in_usd',ascending=False).head()
```

## 4. Какое процентное соотношение каждого региона по вакансиям от всех вакансий?

```python
data.groupby(['company_location'])[['company_location']].count()/len(data)*100
```

## 5. Сколько должностей в наборе данных?

```python
 data['job_title'].value_counts()
```
