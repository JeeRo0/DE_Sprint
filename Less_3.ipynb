## 1. Построение распределений

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import rayleigh as ray
```

```python
x = 0
s = 1
rv = ray(loc=x, scale=s)
r = rv.rvs(size=1000)
```

```python
a = np.linspace(0, 4, 1000)
pdf = rv.pdf(a)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 2, 1])
plt.hist(r, bins=30, color='#10AEE3', label = 'гистограмма выборки', density=True)
plt.plot(a, pdf, color='#F20C3E', label = 'теоретическая плотность')
plt.title('гистограмма выборки и теоретическая плотность вероятности')
plt.xlabel('$ x $')
plt.ylabel('$ f(x) $')
grid = plt.grid(True)  
plt.legend(loc = 1)
```

## 2. Построение графика на по функции

```python
x = np.arange(-1.5,1.5,0.0001)
y1= lambda x: ((x**2)**(1/3)+np.sqrt((x**4)**1/3-4*(x**2-1)))/2
y2= lambda x: ((x**2)**(1/3)-np.sqrt((x**4)**1/3-4*(x**2-1)))/2

plt.plot(x,y1(x))
plt.plot(x,y2(x))
plt.show()
```

## 3.Анализ зависимостей

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charger = []
```

```python
with open("data_home.txt", mode = "r") as file:
    for people in file.readlines():
       
        age.append(int(people.split(",")[0]))
        
        if people.split(",")[1] == "male":
            sex.append(1)
        else:
            sex.append(0)
        
        bmi.append(float(people.split(",")[2]))
        children.append(int(people.split(",")[3]))
        
        if people.split(",")[4] == "yes":
            smoker.append(1)
        else:
            smoker.append(0)
        
        if people.split(",")[5] == "northwest":
            region.append(0)
        elif people.split(",")[5] == "northeast":
            region.append(1)
        elif people.split(",")[5] == "southwest":
            region.append(2)
        elif people.split(",")[5] == "southeast":
            region.append(3)
        else:
            region.append(people.split(",")[5])

        
        charger.append(float(people.split(",")[6]))
```

```python
age = np.array(age)
sex = np.array(sex)
bmi = np.array(bmi)
children = np.array(children)
smoker = np.array(smoker)
region = np.array(region)
charges= np.array(charger)
```
